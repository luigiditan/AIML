# **Machine Learning Algorithm**


**Linear Regression**

Linear Regression is the identified algorithm or method used in performing a machine learning activity given the dataset about (dataset name). Linear Regression is defined as attempting to model the relationship between the two variables by fitting the linear equation to observe the data. The two variables are considered to be an explanatory variable and dependent variable which is the value that will predict using or basing it from the independent variable. It fits in a straight line or surface that minimizes the discrepancies between the predicted values and actual output values.






##***Papers / Publications / White Papers***

According to this study, Linear Regression is a statistical test applied into a dataset to define and quantify the relation between variables. Regression analysis allows the prediction of a dependent variable,s value. The significance of Linear Regression is to analyze the strength of the association between outcome and predictor variables an, adjusts for the effect of covariates or which they called “cofounders”, helps estimate the important risk factors that affect the dependent variable, and Linear Regression also helps analyze the extent of change in independent variable and helps quantify new cases.

**2. Linear Regression Analysis: Theory and Computing (Yan & Su,2009)**

According to this book, Regression is already used 100 years ago, and it was first introduced by Francis Galton in 1908, a renowned British biologist. This term we called “Regression” describe the statistical relations between variables. There are two types of Linear Regression mentioned in the book which is Simple and Multiple Linear Regression. There is a principle called Least Squares Principle for simple linear regression. The function of this principle is to find the estimates b0 and b1 such that the sum of the squared distance from actual response, Yi and predicted response reaches minimum among all possible choices of regression coefficients. The motivation behind the least squares method is to find parameter estimates by choosing the closest regression line to all data points.

\\

**3. Multiple Linear Regression: 2nd Edition (Trammer, Murphy, et. al, 2020)**

According to this paper, Linear regression is a modelling technique used for data analysis to make predictions. There are two types of Linear Regression mentioned in this paper in simple Linear Regression a bivariate model is built to predict a response variable (y) from an explanatory variable represented in x^1 while Multiple Linear Regression is an extended model to include more than one explanatory variable to produce a multivariate model. This paper presents the relevant theory and gives an outline for bivariate and multivariate regression models.


Linear Regression line has a given equation:

$$ Y = M + BX $$

Where:

X is the explanatory variable or independent variable
Y is the dependent variable or the predicted values
B is the slope of line
M is the Y intercept (value of Y when X = 0)

\\
In finding the slope of the regression line, the formula for the slope, m of the best fitting line is 

$$ M = r \ \ (\frac{S_y}{S_x}) $$

Where:

R is the correlation between the values of X and Y
Sy and Sx are the standard deviations of the x values and y values 

\\

In finding the y- intercept of a regression line, the formula for the y intercept, b of the best fitting line is 

$$ B = \tilde{Y} - M \tilde{X} $$

Where:

x̅ and y̅ are the means of the x and y values 
m is the slope

```python
class LinearRegression1:
  def __init__ (self, X, y, X_test, y_test, X_train):
    self.X = X
    self.y = y
    self.X_test = X_test
    self.y_test = y_test
    self.X_train = X_train 

  def train(self):
    self.y = self.y.values
    a = self.X.iloc[::,1].values

    SD_x = np.std(a)
    SD_y = np.std(y)

    r = np.corrcoef(a, self.y)[0,1]

    self.w1 = r * (SD_y/SD_x)
    self.w0 = self.y.mean() - self.w1 * a.mean()
    return print("w0: ", self.w0, "\nw1: ", self.w1)

  def predict(self):
    self.pred = self.w0 + self.w1 * self.X_train
    index = np.array(list(range(0, self.pred.size)))
    return print("y_hat = ", self.pred, ", i = ", index)

  def evaluate(self): # note: y_hat = predicted value; y = orig value; y_bar = y mean
    SSE = np.sum(np.square(self.pred - self.y[0:16598]))
    SST = np.sum(np.square(self.y - self.y.mean()))
    MSE = np.mean(np.square(self.pred - self.y[0:16598]))
    r2 = 1 - (SSE/SST)
    n = self.y.size 
    k = 1
    df = {'SST' : SST,
         'SSR' : np.sum(np.square(self.pred - self.y.mean())),
         'SSE' : SSE,
         'MSE' : MSE,
         'RMSE' : np.sqrt(MSE),
         'R-squared' : r2,
         'Adjusted R-squared' : 1 - ((1 - r2)*(n - 1)/(n-k-1))}
    return df
  
  def visualize(self):
    plt.figure(figsize=(5,5))
    plt.grid()
    plt.scatter(self.X_train, self.y)
    plt.plot(self.X_train, self.pred, color='red')
    return plt.show()
```

```python
class init_LR:
  def __init__ (self, X, y, mini, maxi, steps):
    self.X = X
    self.y = y
    self.mini = mini
    self.maxi = maxi
    self.steps = steps
    self.theta = 0 
    self.b = 1
    self.iter1 = 0
    self.iters = np.array(np.arange(self.mini, self.maxi + self.steps, self.steps))
    self.h = self.theta * self.X + self.b

class LR(init_LR):
  def __init__(self):
    init_LR.__init__ (self, X, y, mini, maxi, steps)
  
  def SSE(self):
    self.j = np.sum(np.square(self.h-self.y))
    return self.j
  
  def optimization(self):
    ## create logs for optimization backtracking
    history = {
        'epoch': [],
        'theta': [],
        'model': [],
        'loss': []
    }

    for iter in self.iters:
      ## set the angle as the normalized ratio of the iter to max
      self.theta = iter/self.maxi
      ## compute for the hypothesis
      self.h = self.theta * self.X + self.b
      ## compute for the loss
      self.J = np.sum(np.square(self.h-self.y))

      ## log the changes
      history['epoch'].append(iter)
      history['theta'].append(self.theta)
      history['model'].append(self.h)
      history['loss'].append(self.J)
      self.hists_df = pd.DataFrame(history)
    return self.hists_df
  
  def graph_loss(self):
    plt.xlabel('Theta')
    plt.ylabel('Loss')
    return plt.plot(self.hists_df['theta'], self.hists_df['loss'])
```
